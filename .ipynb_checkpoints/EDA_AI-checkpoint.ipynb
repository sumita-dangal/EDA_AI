```python
# Load the CSV file into a DataFrame
file_path = './diabetes.csv'  # Replace 'your_file_path.csv' with the actual file path
data = pd.read_csv(file_path)

# Filter rows where 'Pregnancies' is less than or equal to 5
filtered_data = data[data['Pregnancies'] <= 5]

# Save the filtered DataFrame back to a CSV file
filtered_data.to_csv('filtered_data.csv', index=False)

print("Data filtered and saved successfully!")
```

```python
data = pd.read_csv("./filtered_data.csv")
```

```python
data.head(5)
```

```python
data.shape
```

```python
data['Outcome'][data['Outcome'] == 0] = 'Healthy'
data['Outcome'][data['Outcome'] == 1] = 'Diabetes'
```

```python
# Categorical data
# countplot
sns.countplot(data=data, x='Outcome')
data['BMI'].value_counts()
```

```python
data['Outcome'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Histogram treat as categorical data distribution of data

plt.hist(data['Age'])
```

```python
# Distplot 
# probability density function --> The line is kde 
# skweness 
sns.distplot(data['DiabetesPedigreeFunction'])
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column 
sns.scatterplot(x='Age', y='DiabetesPedigreeFunction',hue='BMI', style='Pregnancies',data=data) 
plt.show()
```

```python
# Bar plot (Numerical - Categorical) 

sns.barplot(x='Age',y='BMI',hue='Insulin',data=data)
```

```python
sns.boxplot(data['DiabetesPedigreeFunction'])
```

```python
sns.lineplot(x='Age',y='DiabetesPedigreeFunction',data=data)
```

```python
sns.heatmap(data.pivot_table(values='Age',index='DiabetesPedigreeFunction',columns='Pregnancies'))
```

```python
# pairplot 
sns.pairplot(data,hue='Age')
```

```python
# pairplot 
sns.pairplot(data,hue='Outcome')
```
